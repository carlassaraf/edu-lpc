```python
import matplotlib.pyplot as plt
import numpy as np

f = np.polyfit([1, 1000], [250, 100], 1)

frequencies = [i for i in range(1, 1001)]
amplitude = np.polyval(f, frequencies)

plt.title("Amplitude multiplier vs frequency")
plt.ylabel("Amplitude multiplier")
plt.xlabel("Frequency [Hz]")

plt.grid()
plt.plot(frequencies, amplitude, label=f"{f[0]:.2f}x + {f[1]:.2f}")
plt.legend()
plt.show()
```

```python
f = np.polyfit([1, 1000], [100, 10], 1)

frequencies = [i for i in range(1, 361)]
points = np.polyval(f, frequencies)

plt.title("Number of points vs frequency")
plt.ylabel("Number of points")
plt.xlabel("Frequency [Hz]")

plt.grid()
plt.plot(frequencies, points, label=f"{f[0]:.2f}x + {f[1]:.2f}")
plt.legend()
plt.show()
```
